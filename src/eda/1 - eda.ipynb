```python
import os
import pandas as pd
import pyodbc

```

```python
from dotenv import load_dotenv

load_dotenv()

server_name = os.environ.get("server_name")
database = os.environ.get("database")
```

```python
query = """
SELECT TOP (1000) [POLICYSTATUS]  
      ,[UNIQUE_CLAIMS]
      ,[INVOICEBENEFIT]
      ,[INVOICELINEDETAILS]
      ,[INVOICEID]
      ,[INVLINEPBCDISCOUNTAMOUNT]
      ,[INVLINEPBCDENIEDAMOUNT]
      ,[BENEFITEXCESSAPPLIED]
      ,[INVOICEAMOUNT]
      ,[INVOICEDAMOUNT]
      ,[SETTLEDAMOUNT]
      ,[INVOICELINEUSERSTATUS]
      ,[PAYMENT_AMOUNT]
      ,[PRODUCT]
      ,[INVOICELINESTATUS]
      ,[INVOICESTATUS]
      ,[Claims_Invoice_Status]
      ,[Claim_Decline_Status]
      ,[INVOICECURRENCYDENIEDAMOUNT]
      ,[CAPTUREDATE]
      ,[YR]
      ,[ASSESSEDDATE]
      ,[CAPTUREDBY]
      ,[ASSESSEDBY]
      ,[ReasonText]
      ,[FIRSTDIAGNOSIS]
      ,[SECONDDIAGNOSIS]
      ,[ICD10]
      ,[DELIVERYCATEGORY]
      ,[PAYMENTSTATUSDESC]
	  ,[INVOICELINESUBBENEFIT]
      ,[CATEGORYNAME]
      ,[RELATIONSHIP]
      ,[CHARGECATEGORY]
      ,[QA_COMMENTS]
      ,[AUTHENTICATION_TYPE]
	  ,[CURACEL_DECLINE_REASON]
      ,[CURACEL_STATUS]
      ,[CURACEL_AMOUNT]
FROM [Jubilee].[dbo].[ClaimsRegister]  -- Replace with your table name
"""
```

```python
conn_str = (
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=' + server_name + ';'
    'DATABASE=' + database + ';'
    'Trusted_Connection=yes;'
)

# Connect to the database and execute the query
conn = pyodbc.connect(conn_str)
data = pd.read_sql(query, conn)
conn.close()
```
